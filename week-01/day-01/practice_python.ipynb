```python
# VARIABLES

name = "Manish"
print(f"Hello, {name}!")
```

```python
# ARRAYS/LISTS

tools = ["AI", "Web Dev", "App Dev"]
tools.append("UI designing")
first = tools[0]
last = tools[-1]
subset = tools[1:3]

print(f"tools, {tools}")
print(f"First: {first}, Last: {last}")
print(f"Subset, {subset}")
```

```python
# DICTIONARIES

courses = {
  "topic": "AI",
  "length": "2.5",
  "type": "active learning"
}

print(f"Course topic, {courses["topic"]}")
courses["is_interactive"] = "True"
print(f"Courses, {courses}")
```

```python
# FUNCTIONS

def say_hello(name):
  return f"Hello, {name}"

def say_hello_with_title(name, title="Mr"):
  return f"Hello, {title}. {name}!"

print(say_hello("Manish"))
print(say_hello_with_title("Yuvi", "Dr"))
```

```python
# LOOPS
tools = ["AI", "Web Dev", "App Dev"]
for tool in tools:
  print(f"Exploring {tool}")

for i, tool in enumerate(tools):
  print(f"{i}, {tool}")


# LIST COMPREHENSION
upper_tools = [t.upper() for t in tools]
print(f"Upper: {upper_tools}")
```

```python
# CLASS
class Agent:
  def __init__(self, name, model):
    self.name = name
    self.model = model

  def respond(self, question):
    return f"{self.model}'s {self.name} thinks about: {question}"

agent = Agent("MyBot", "gpt-4o-mini")
print(agent.respond("What is AI?"))
```
